# Imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Create Plot Plane

```python
xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

## Plot with Legend

```python
plt.plot(xs, ys)
```

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$') # assign label
plt.legend()                                  # add legend
```

## Combined Plots

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.sin(xs), label='sin(x)')       # add new plot with label
plt.legend()
```

## Parallel Plots

```python
# we want 2 plots in 1 column, and select plot 1
plt.subplot(2, 1, 1)
plt.plot(xs, np.sin(xs), label='sin')
plt.legend()
# we want 2 plots in 1 column, and select plot 2
plt.subplot(2, 1, 2)
plt.fill(xs, -xs**2, label=r'${-x}^2$')
plt.legend()
```

## Plots in Grids

```python
# we want 2 plots in 2 column, and select plot 1
plt.subplot(2, 2, 1)
plt.plot(xs, np.sin(xs), label='sin')
plt.legend()
# we want 2 plots in 2 column, and select plot 2
plt.subplot(2, 2, 2)
plt.fill(xs, -xs**2, label=r'${-x}^2$')
plt.legend()
# we want 2 plots in 2 column, and select plot 3
plt.subplot(2, 2, 3)
plt.fill(xs, -xs**2, label=r'${-x}^2$')
plt.legend()
# we want 2 plots in 2 column, and select plot 4
plt.subplot(2, 2, 4)
plt.plot(xs, np.sin(xs), label='sin')
plt.legend()
```
